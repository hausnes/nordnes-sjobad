```python
import matplotlib.pyplot as plt
import csv

# Initialiser lister for  lagre data
tid = []
temperatur = []
lufttemperatur = []

# Les data fra CSV-fil
with open('temperatur.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    next(plots, None)  # Skip the headers
    for row in plots:
        if row[1] != '-' and row[2] != '-':
            tid.append(row[0][5:16])
            temperatur.append(float(row[1]))
            lufttemperatur.append(float(row[2]))

# Lag en graf
plt.plot(tid, temperatur, label='Sjtemperatur')
plt.plot(tid, lufttemperatur, label='Lufttemperatur')

# Legg til titler og etiketter
plt.xlabel('Tid')
plt.ylabel('Temperatur')
plt.title('Sj- og lufttemperatur over tid')
plt.legend()

# Sett x-aksen etiketter til  rotere 90 grader
plt.xticks(rotation=90)

# Vis bare et utvalg av tidspunkt
plt.xticks(tid[::10])  # Change the 10 to any number that gives you the desired number of labels

# Vis grafen
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

# Initialiser lister for  lagre data
temperatur = []
lufttemperatur = []

# Les data fra CSV-fil
with open('temperatur.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    next(plots, None)  # Skip the headers
    for row in plots:
        temperatur.append(float(row[1]))
        lufttemperatur.append(float(row[2]))

# Lag et scatterplot
plt.scatter(temperatur, lufttemperatur)

# Legg til titler og etiketter
plt.xlabel('Sjtemperatur')
plt.ylabel('Lufttemperatur')
plt.title('Sammenheng mellom sj- og lufttemperatur')

# Vis grafen
plt.show()
```

```python
'''
En korrelasjonskoeffisient nr 1 indikerer en sterk positiv sammenheng, 
en nr -1 indikerer en sterk negativ sammenheng, 
og en nr 0 indikerer ingen sammenheng.
'''

import matplotlib.pyplot as plt
import csv
import numpy as np

# Initialiser lister for  lagre data
temperatur = []
lufttemperatur = []

# Les data fra CSV-fil
with open('temperatur.csv', 'r') as csvfile:
    plots= csv.reader(csvfile, delimiter=',')
    next(plots, None)  # Skip the headers
    for row in plots:
        temp = float(row[1]) if row[1] != 'NaN' else None
        lufttemp = float(row[2]) if row[2] != 'NaN' else None
        if temp is not None and lufttemp is not None:
            temperatur.append(temp)
            lufttemperatur.append(lufttemp)

# Beregn korrelasjonskoeffisienten
corr = np.corrcoef(temperatur, lufttemperatur)[0, 1]
print(f'Korrelasjonskoeffisienten er {corr:.2f}')

# Lag et scatterplot
plt.scatter(temperatur, lufttemperatur)

# Legg til titler og etiketter
plt.xlabel('Sjtemperatur')
plt.ylabel('Lufttemperatur')
plt.title('Sammenheng mellom sj- og lufttemperatur')

# Vis grafen
plt.show()
```
